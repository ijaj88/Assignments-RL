## 2. Explore ‘CartPole-v1’

```python



env = gym.make("CartPole-v1", render_mode="rgb_array")


observation, info = env.reset()

episodes =1000
reward_perepisode=np.zeros(episodes)
stepdata=[]


for i in range(episodes):
    observation, info = env.reset()
    terminated = False
    truncated = False
    cumulative_reward=0
    while not terminated :
        action=np.random.choice(env.action_space.n)
        
        observation, reward, terminated, truncated, info = env.step(action)
        #print(reward)
        cumulative_reward+=reward
        stepdata.append(reward)

        #envrend=env.render()
        #plt.imshow(envrend)
        #clear_output(wait=True)
        #plt.show()

    reward_perepisode[i] =  cumulative_reward  
    
envrend=env.render()
plt.imshow(envrend)
plt.show()         
env.close()
```

### Rendering the cart pole 

```python
plt.subplot(1,2,1)
plt.plot(reward_perepisode)
plt.ylabel('Rewards Per episode')
plt.xlabel('Episode #')

plt.subplot(1,2,2)
plt.plot(stepdata)
plt.ylabel('Rewards Per steps')
plt.xlabel('steps #')
```

## Describing the CartPole envirorment

A pole is attached by an un-actuated joint to a cart, which moves along a frictionless track. The pendulum is placed upright on the cart and the goal is to balance the pole by applying forces in the left and right direction on the cart. 


Action: It has two action 0/1 to push to left and right respectively


Observation: It has 4 observation  (a) Cart position (b) Cart velocity (c) Pole Angle and (d) Pole Angular velocity

Reward:  a reward of +1 for every step taken, including the termination step, is allotted. For every step 1 reward is added till the termination or truncation ocuured. The idea is to remain in stable position as long as possible.


As a result in each episode, the cumulative reward is equal to he duration of the episode. 

## 3. Explore Acrobot-v1

### Describing the acrobot environment
The system consists of two links connected linearly to form a chain, with one end of the chain fixed. The joint between the two links is actuated. The goal is to apply torques on the actuated joint to swing the free end of the linear chain above a given height while starting from the initial state of hanging downwards.

 Action: The action space is three  and represents the torque applied on the actuated joint between the two links.0,1 and 2 respresents -1 N, 0 N and 1 N respectively torque applided at the joint

 Observation :The observation is a ndarray with shape (6,) that provides information about the two rotational joint angles as well as their angular velocities


 Rewards : the goal is to have the free end reach a designated target height in as few steps as possible, and as such all steps that do not reach the goal incur a reward of -1. Achieving the target height results in termination with a reward of 0.

 The negative ensures that the target is achived in minimum number of steps . At each step until it reaches the terminal state -1 is penalised and for each episodes the cumulative reward is negative of number of steps plus 1. The last step is awarded no penally
